##### Missing values in Donor_Age. Find missing value using median age

```python
dataset['DONOR_AGE'].median()
```

```python
#find null age values and fill in missing values 
def impute_age(cols):
    Age = cols[0]
    
    if pd.isnull(Age):
        return 60
    else:
        return Age
```

```python
##complete missing age fields

dataset['DONOR_AGE'] = dataset[['DONOR_AGE']].apply(impute_age,axis=1)
```

##### Missing values in Donor_Age. Find missing value using median age

```python
dataset['INCOME_GROUP'].median()
```

```python
#find null income group values and fill in missing values 

def impute_incgroup(cols):
    incgroup = cols[0]
    
    if pd.isnull(incgroup):
        return 4
    else:
        return incgroup
```

```python
dataset['INCOME_GROUP'] = dataset[['INCOME_GROUP']].apply(impute_incgroup,axis=1)
```

##### Missing values in Wealth_rating. Find missing values based on their income group

```python
plt.figure(figsize=(12, 7))
boxes = sns.boxplot(x='INCOME_GROUP',y='WEALTH_RATING',data=dataset, medianprops={'linewidth':2}, color='yellow')

plt.title('Wealth based on Income Group')
plt.show()
```

```python
# Due to the relationship seen above, we can impute missing wealth_rating based on Income group, and using 
#the mean values for each group from the plot above

def impute_wlthrat(cols):
    wlthrat = cols[0]
    incgroup = cols[1]
    
    if pd.isnull(wlthrat):

        if incgroup == 1:
            return 3

        elif incgroup == 2:
            return 4
        
        elif incgroup == 3:
            return 5
        
        elif incgroup == 4:
            return 5
        
        elif incgroup == 5:
            return 6
        
        elif incgroup == 6:
            return 7
        
        else:
            return 8

    else:
        return wlthrat
```

```python
dataset['WEALTH_RATING'] = dataset[['WEALTH_RATING','INCOME_GROUP']].apply(impute_wlthrat,axis=1)
```
