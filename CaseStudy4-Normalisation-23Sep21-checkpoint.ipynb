#### Scaling (Standardisation)

scaling is used to ensure values in each column are all in the 
same "ballpark" range, typically between -1 and 1

```python
df.dtypes # check data types
```

```python
df.info() # shows data types for all columns
```

```python
# Exercise

# a) isolate the numerical columns (so we scale the numerical data)
# b) drop the date-like features
# c) identify any other columns we may not need to scale
```

```python
# a) isolate the numerical columns
dfNum = df.select_dtypes(exclude = ["object"])
dfNum.columns
```

```python
len(dfNum.columns)
```

```python
# b) drop the date-like features
# c) identify any other columns we may not need to scale
dfNum = dfNum.drop(['YearBuilt', 
                    'YearRemodAdd',
                    'GarageYrBlt',
                    'MoSold', 
                    'YrSold',
                    'Id',
                     'SalePrice'], axis =1)

dfNum.head() # quick look
```

we also drop Id as it has no "predictive power" in an ML model and 
SalePrice as it is NOT a feature (SalePrice is the target)

```python
dfNum.columns
```

```python
# scaling often makes the difference b/w an average model and a good one
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # STEP 1 - select the "type" of scaling we want
scaler.fit(dfNum) # STEP 2 - fit the scaling technique to our NUMERICAL data 
scaled_df = scaler.transform(dfNum) # STEP 3 transform our dataframe (i.e. scale it)

# POST-PROCESSING
# NB StandardScaler converts the scaled data to a numpy array, 
# to reverse back to a dataframe, we use pd.DataFrame
scaled_df = pd.DataFrame(scaled_df, columns = dfNum.columns)
scaled_df.head()
```
